### Methods - Simple Feature Scaling, Min-Max, Z-Score/Standard Score

### SIMPLE FEATURE SCALING METHOD

- Mathematically, x(new) = x(old)/x(max)
- Result range: 0 to 1.

### Min-Max Method. 
- x(new) = x(old) - x(min) / x(max) - x(min)
- Result range: 0 to 1.

### Z-Score / Standard Score Mathod
- Mathematically, x(new) = x(old) - mu / sigma
- mu :- Average of feature
- sigma :- Standard Deviation
- Result range: -3 to +3.
