```python
#### POLYNOMIAL REGRESSION - They are useful for curvilinear relations.
# Y = b0 + b1x1 + b2(x1)^2 --> QUADRATIC
# Y = b0 + b1x1 + b2(x1)^2 + b3(x1)^3 --> CUBIC
# Y = b0 + b1x1 + b2(x1)^2 + b3(x1)^3..... --> HIGHER ORDER
```

```python
import numpy as np
import seaborn as sns
flights = sns.load_dataset("flights")
x = np.random.randn(100)
y = np.random.randn(100)
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
```
