```python
import pandas as pd
path = "C:/Users/Shri/Coding/Additional-Files/data-to-learn-import.xlsx"
df = pd.read_excel(path)
print(df)
```

```python
import numpy as np
grp = df.groupby("A")
grp.aggregate(np.sum)
```

```python
grp = df.groupby(["A","B"])
grp.aggregate(np.sum)
```

```python
# Spaces are adjusted.
grp = df.groupby(["A","B"], as_index = False)
grp.aggregate(np.sum)
```
