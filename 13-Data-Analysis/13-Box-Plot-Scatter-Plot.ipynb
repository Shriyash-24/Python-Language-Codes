### BOX PLOT AND SCATTER PLOT

```python
import pandas as pd
path = "C:/Users/Shri/Coding/Additional-Files/data-to-learn-import.xlsx"
df = pd.read_excel(path)
print(df)
```

```python
#### BOXPLOT
import seaborn as sns
sns.boxplot(x = "Length", y = "E", data = df)
```

```python
#### SCATTERPLOT
import matplotlib.pyplot as plt
x = df["Length"]
y = df["E"]
plt.scatter(x,y)
plt.title("Scatterplot along the X-Y Axis.")
plt.xlabel = "Length"
plt.ylabel = "Width"
```
