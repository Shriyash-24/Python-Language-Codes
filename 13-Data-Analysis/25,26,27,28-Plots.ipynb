```python
import pandas as pd
path = "C:/Users/Shri/Coding/Additional-Files/data-to-learn-import.xlsx"
df = pd.read_excel(path)
import seaborn as sns
flights = sns.load_dataset("flights")
```

#### REGRESSION PLOTS

```python
# The vertical axis is dependent and horizontal axis is independent.
import matplotlib.pyplot as plt
sns.regplot(x="year", y="passengers",data = flights)
plt.ylim(0,)
```

```python
## RESIDUAL PLOT - USED TO FIND PROBLEMS WITH REGRESSION.
## RESIDUAL VALUE = IT IS A MEASURE OF HOW MUCH A REGRESSION VERTICALLY MISSES A DATA POINT.
```

```python
sns.residplot(data = flights, x = "year", y = "passengers")
```

```python
### DISTRIBUTION PLOT - It will display numeric values which are plotted against a dimension.
import seaborn as sns
import numpy as np
np.random.seed(0)
x = np.random.randn(100)
ax_1 = sns.distplot(x)
```

```python
ax_2 = sns.distplot(x, rug = True, hist = False)
```

```python
ax_2 = sns.distplot(x, rug = False, hist = True)
```

```python
from scipy.stats import norm
ax_3 = sns.distplot(x, fit = norm)
```

```python
ax_3 = sns.distplot(x, fit = norm, kde = False) # kde is kernal density estimation, represent data in single density curve.
```

```python
ax_4 = sns.distplot(x, vertical = True)
```

```python
sns.set_color_codes()
ax_5 = sns.distplot(x, color = 'r')
```

```python
sns.set(style = 'dark')
ax_6 = sns.distplot(x, color = 'g')
```
