```python
import pandas as pd
path = "C:/Users/Shri/Coding/Additional-Files/data-to-learn-import.xlsx"
df = pd.read_excel(path)
print(df)
```

```python
# Numeric To Categorical
import numpy as np
bins = np.linspace(min(df["Length"]),max(df["Length"]),4) # linspace is used to determine evenly spaced sequence in  a specificied interval
# start, stop, no of samples.
```

```python
group_names = ["Low","Medium","High"]
```

```python
df["Length-binned"] = pd.cut(df["Length"],bins,labels = group_names, include_lowest = True)
```

```python
# Categorical To Numeric - One Hot Encoding
pd.get_dummies(df["Length"])
```
