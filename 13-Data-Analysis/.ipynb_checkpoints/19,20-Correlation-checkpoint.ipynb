```python
import seaborn as sns
flights = sns.load_dataset("flights")
sns.regplot(x="passengers", y = "year", data = flights)
```

```python
# This is a positive correlation.
```

- Correlation - Measures to what extent different variables are interdependent.
- Eg - Lung Cancer --> Smoking.

- 1. Pearson Correlation - Correlation corefficient, P-value.

Correlation coefficient determines strength of relation between variables. 
- Close to +1 - Large Positive Relationship
- Close to -1 - Large Negative Relationship
- Close to 0 - No Relationship

P-value also determines strength and checks for significance.
- Value <0.001 - STRONG
- Value <0.05 - Moderate
- Value <0.1 - Weak
- Value >0.1 - No Certainty.

### STRONG CORRELATION --> Correlation co-efficient is close to +1 & -1 and P-Value is <0.001.

```python
from scipy import stats
flights = sns.load_dataset("flights")
print(flights)
```

```python
pearson_correlation = stats.pearsonr(flights['year'], flights['passengers'])
print(pearson_correlation)
```

```python
# So, Corelation coefficient is close 1 --> STRONG, p-value is < 0.01 (notice 10^-60) --> STRONG.
```
