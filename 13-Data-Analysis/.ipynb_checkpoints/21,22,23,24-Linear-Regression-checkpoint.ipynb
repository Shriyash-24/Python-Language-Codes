### LINEAR REGRESSION
#### SIMPLE LINEAR REGRESSION

- Modelling the relationship from data.
- 2 types - Simple linear regression and multiple linear regression.
- We will have independent variable (x1) --> Linear Regression --> Predicts the value.
- PREDICTOR (independent) Variable - x
- TARGET (dependent) Variabel - y
- y = b0 + b1x --> b0 = intercept, b1 = slope.

#### MULTIPLE LINEAR REGRESSION 
- Multiple Independent Variables (x1..xn) --> Linear Regression --> Predicts the value.
- PREDICTOR (independent) Variable - x1,x2,x3..
- TARGET (dependent) Variable - y
- y = b0 + b1x1 + b2x2 + b3x3...bnxn. --> b0 = intercept, b1 = parameter of x1, b2 = parameter of x2.

```python
# SIMPLE LINEAR REGRESSION USING PYTHON
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
import seaborn as sns
flights = sns.load_dataset("flights")
x = flights[['year']]
y = flights['passengers']
```

```python
lm.fit(x,y)
```

```python
Yhat = lm.predict(x)
print(Yhat)
```

```python
lm.intercept_
```

```python
lm.coef_
```
