```python
### MEASURES FOR IN-SAMPLE EVALUATION
### PARAMETERS - MSE (MEAN SQUARE ERROR) & R-SQUARED (R^2)
```

- MSE - Avg between estimated and actual value
- R^2 - How close data is fitted?

- Subtracting - y = y1 - y (cap)1
- Squaring - y^2
- Mean/Average of all the errors by adding them all together and dividing by number of samples.
- Mean = Adding all of them / No. Of Samples

```python
from sklearn.metrics import mean_squared_error
import seaborn as sns
flights = sns.load_dataset("flights")
```

```python
mean_squared_error(flights['year'], flights['passengers'])
```

```python
mean_squared_error(flights['year'], flights['passengers'], multioutput = "raw_values") # raw values enters full set of errors.
```

```python
mean_squared_error(flights['year'], flights['passengers'], multioutput = "uniform_average")
```

```python
mean_squared_error(flights['year'], flights['passengers'], squared = True)
```

```python
mean_squared_error(flights['year'], flights['passengers'], squared = False) # It returns the root mean squared error.
```

- If the ration is 1, then we get 0, the line is a bad fit for the data.
