### Binning - Grouping & Conversion. 

#### Normal Data

| S.NO | PRICE |
|------|-------|
|  1   |  1000 |
|  2   |  600  |
|  3   |  500  |
|  4   |  100  |

#### Binning - Converts numerical data into categorical data

| S.NO | PRICE | PRICE BINNED |
|------|-------|--------------|
|  1   |  1000 |    HIGH      |
|  2   |  600  |    MEDIUM    |
|  3   |  500  |    MEDIUM    |
|  4   |  100  |    LOW       | 

#### You can also convert data from categorical data to numerical data. It is known as One Hot Encoding.

#### Normal Data
| S.NO | FUEL     |
|------|----------|
|  1   | GASOLINE |
|  2   | DIESEL   |
|  3   | DIESEL   | 
|  4   | GASOLINE |

#### One Hot Encoding
| S.NO | FUEL     | GASOLINE | DIESEL |
|------|----------|----------|--------|
|  1   | GASOLINE |   1      |   0    |
|  2   | DIESEL   |   0      |   1    |
|  3   | DIESEL   |   0      |   1    | 
|  4   | GASOLINE |   1      |   0    |
